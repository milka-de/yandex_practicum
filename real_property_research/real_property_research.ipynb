## Предобработка данных

### Стиль заголовка

```python
df = df.rename(columns={'cityCenters_nearest': 'city_centers_nearest'})
display(df.columns)
```

Название колонки cityCenters_nearest — расхождения с хорошим стилем, меняем его на более удобное

### Пропуски

```python

display(df.isna().sum())
```

Больше половины столбцов имеют пропуски, к каждому из них нужен свой подход:

* `ceiling_height`
* `living_area`                   
* `kitchen_area`   
* `city_centers_nearest`

Эти столбцы мы заменим на медиану, так как нулевого значения у них быть не может и NaN нам помешает при дальнейшей работе с ними

Но сделаем это не в данном разделе, а в разделе, где избавляемся от аномальных данных. Если мы сделаем это сейчас, то эти данные могут быть искажены

Остальные пропуски заменим сейчас

такие столбцы как:
* `ponds_nearest`          
* `airports_nearest`                      
* `parks_nearest` 

мы не будем трогать, так как они относятся к тем значениям, которых может и не быть. Т.е. недвижимость без потолков найти просто нереально - это нечто уже явно ею считаться не будет, а вот недвижимость у которой поблизости нет парка вполне имеет место быть - вот и причина пропусков

```python
df['is_apartment'] = df['is_apartment'].fillna(0)
#df['days_exposition']= df['days_exposition'].fillna(0)          
df['balcony']= df['balcony'].fillna(0)
df['parks_around3000']= df['parks_around3000'].fillna(0) 
df['ponds_around3000']= df['ponds_around3000'].fillna(0) 
```

`is_apartment` в этом столбце так много пропусков вместо значения - "0", которое в булевом значении означает False - т.е. это не аппартаменты

`balcony`, `parks_around3000`, `ponds_around3000` - в этих  столбцах пропуски вместо значения - "0", но уже просто как число - балконов может в квартире и не быть и объявление с сайто могло уйти в первые же сутки - такое случается не частно, но вариант возможен

`days_exposition` пропусков в этом столбце слишком много, заменив их можем исказить анализ, причину их появления определённо необходимо уточнять у заказчика

```python
df['floors_total']= df['floors_total'].fillna(1)
```

`floors_total` так же как и в предыдущих двух столбцах, скорее всего кол-во не указали, так как этажей больше нет в доме, но в этом случае меняем на 1, мы всё-таки не в Европе, нулевого этажа у нас нет

```python
df['locality_name']= df['locality_name'].fillna('Санкт-Петербург')
```

`locality_name` можно предположить, что, так как данные строго по Питерской недвижимости и облости, люди могли предпочесть не заполнять эту колонку, если их недвижимость находится в Питере

### Аномальные данные

В изучении данных мы использовали метод `.describe()` что позволило нам изучить имеют ли столбцы выбросы

Если брать в расчёт, что Питер славится своими высокими потоками, благодоря которому можно создать двухэтажную квартиру, а также студии, которые так полюбились молодым студентам, то можно объяснить такие минимальные значение по площадям, но вот потолок длиной в 100 метров - это уже перебор

Ограничим этот диапазон 20 метрами, так как остальное уже похоже на выбросы

```python
display('количество выбрасов до удаления', df[df['ceiling_height'] > 20].groupby('ceiling_height')['ceiling_height'].count())

df['ceiling_height'] = df['ceiling_height'].where(df['ceiling_height'] < 20, df['ceiling_height'] / 10 )

display('количество выбрасов после', df[df['ceiling_height'] > 20]['ceiling_height'].count())
```

Теперь мы можем вернуться к удалению пропусков в столбцах `ceiling_height`, `living_area`, `kitchen_area`, `city_centers_nearest`


```python
columns_to_replace = ['ceiling_height', 'living_area', 'kitchen_area', 'city_centers_nearest']
for column in columns_to_replace:
    df[column] = df[column].fillna(df[column].median())
display(df.isna().sum())    
```

### Изменение типов данных

```python
df['last_price'] = df['last_price'].astype('int')
df['floors_total'] = df['floors_total'].astype('int')
df['balcony'] = df['balcony'].astype('int')
df['days_exposition'] = df['days_exposition'].astype('float32').astype('Int64') # закостылила, напрямую в инт не переводит
df['parks_around3000'] = df['parks_around3000'].astype('int')
df['ponds_around3000'] = df['ponds_around3000'].astype('int')

df['total_area'] = df['total_area'].astype('float32')
df['ceiling_height'] = df['ceiling_height'].astype('float32')
df['living_area'] = df['living_area'].astype('float32')
df['kitchen_area'] = df['kitchen_area'].astype('float32')
df['city_centers_nearest'] = df['city_centers_nearest'].astype('float32')
df['airports_nearest'] = df['airports_nearest'].astype('float32')
df['parks_nearest'] = df['parks_nearest'].astype('float32')
df['ponds_nearest'] = df['ponds_nearest'].astype('float32')


df['is_apartment'] = df['is_apartment'].astype('bool')
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
df.info()
```

дату и время привели к правильному типу данных - `datetime64`

`last_price` - привели к типу `int` так как значения довольно большие и подсчёт копеек здесь неуместен

`floors_total`, `balcony`, `days_exposition`, `ponds_around3000`, `parks_around3000` - привели к типу `int` так как они в целом не имеют дробной части       

`is_apartment`  - привели к правильному для столбца типу данных `bool`

столбцы в которых содержатся измерения в м и м2 оставили типом `float` так как не все они имеют высокие показателеи, при которой дробной частью уместно было бы принебречь

### Дубликаты

```python
display(sorted(df['locality_name'].unique()))
display(df.duplicated().sum())
```


 
На этапе изучения данных мы заметили, что данные не имеют дубликатов, а переменная `locality_name` имеет очень много похожих названий за которыми могут скрываться неявные дубликаты

```python
df['locality_name'] = df['locality_name'].str.replace('ё', 'е')
display(df.duplicated().sum())
```

Количество дубликатов в датафрэйме не изменилось, а это значит, что неявных дубликатов не было

### Вывод

В данном этапе мы провели предобработку данных и теперь данные подготовлены к изучению

В данных присутствовали пропуски, аномалии, неподходящие типы данных и вид зааголовка колонок, дубликаты отсутствовали

## Расчёты и добавление результатов в таблицу

```python
filter1 = df['floor'] != 1
filter2 = df['floor'] != df['floors_total']
df['floor_kategory'] = (
    df['floor'].where((filter1 & filter2)== False, 'другой')
    .where(filter1, 'первый')
    .where(filter2, 'последний')
)
```

добавили в датафрейм столбец, который разделяет этажи по трём категориям

```python
df['day'] = pd.DatetimeIndex(df['first_day_exposition']).day
df['month'] = pd.DatetimeIndex(df['first_day_exposition']).month
df['year'] = pd.DatetimeIndex(df['first_day_exposition']).year
```

добавили в датафрейм столбцы с днём, месяцем, годом публикации

```python
df['price_area_m'] = df['last_price'] / df['total_area']
df['price_area_m'] = df['price_area_m'].astype('int')
```

Ищем цену квадратного метра и выписываем в отдельный стольбец

```python
df['living_on_total'] = df['living_area'] / df['total_area']
df['kitchen_on_total'] = df['kitchen_area'] / df['total_area']
```

Ищем соотношение жилой и общей площади и также создаём новые столбцы
